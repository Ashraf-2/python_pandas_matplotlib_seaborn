```python
import matplotlib.pyplot as plt
```

```python
# Data 
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

# Plot
plt.plot(x,y) # line plot
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Simple line plot")
plt.legend(["y = growth"])
# Show plot
plt.show()
```

```python
# add colors and other style
x = [1, 2, 3, 4, 5]  
y = [10, 20, 25, 30, 40]  

plt.plot(x,y, label ="sales data", color='red', linestyle='--', marker='o')
plt.title("Customized Sales Data")  
plt.legend()  

# Setting axis limits  
plt.xlim(0, 6)  
plt.ylim(0, 50)  

# Custom tick marks  
plt.xticks([0, 1, 2, 3, 4, 5, 6])  
plt.yticks([0, 10, 20, 30, 40, 50]) 
```

```python
df_usa = df[df['Country'] =='USA']
df_usa

df_usa_nissan = df[(df['Country'] == 'USA') & (df['Brand'] == 'Nissan')]
df_usa_nissan

df_usa_toyota = df[(df['Country'] == 'USA') & (df['Brand'] == 'Toyota')]
df_usa_toyota
```

```python
x1 = df_usa_nissan['Year']
y1 = df_usa_nissan['Units_Sold']

plt.plot(x1,y1, color='red', marker='x')

x2 = df_usa_toyota['Year']
y2 = df_usa_toyota['Units_Sold']
plt.plot(x2,y2, color='blue', marker='o')   
plt.legend(["Nissan Sales in USA", "Toyota Sales in USA"])
```
