```python
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np
```

## Scannter Plot

```python
# Data  
x = [1, 2, 3, 4, 5, 6]  
y = [10, 15, 8, 20, 25, 18]

plt.scatter(x,y)
plt.xlabel("X-Axis")  
plt.ylabel("Y-Axis")  
plt.title("Basic Scatter Plot")  
```

```python
plt.plot(x,y) #line plot
```

```python
# customizing the plot
plt.scatter(x,y, color='red', marker='*', alpha=0.5, s=100) #s is size of marker
plt.title("Scatter Plot with Custom Markers")  
```

```python
# Data  
x = np.random.rand(50) * 10  
y = np.random.rand(50) * 100 

colors = np.random.rand(50)  # Values for colormap    # Values for colormap  
scatter = plt.scatter(x, y, c=colors, cmap="viridis", s=100, alpha=0.8) 
plt.colorbar(scatter) 

plt.show()  
```

```python
plt.scatter(x,y,cmap='viridis')
plt.show()
```

```python
x = [1, 2, 3, 4, 5]  
y1 = [5, 15, 10, 20, 18]  
y2 = [8, 12, 6, 25, 22] 

plt.scatter(x, y1, label='Data 1') 
plt.scatter(x, y2, label='Data 2')
plt.title("Scatter Plot with Different Markers") 
plt.legend()
```

 Common Marker Styles:

"o"  Circle,
"s"  Square,
"^"  Triangle,
"D"  Diamond,

```python
# Data  
x = [1, 2, 3, 4, 5]  
y = [10, 20, 15, 30, 25]  

# Plot  
plt.scatter(x, y, color="green")  

# Adding annotations  
for i, txt in enumerate(["A", "B", "C", "D", "E"]):  
    plt.annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(5,5), ha='center')


plt.show()  
```

```python
x = [1, 2, 3, 4, 5]  
y = [10, 20, 15, 30, 25] 
c = [15, 25, 35, 45, 55]

plt.scatter(x,y)

for i,txt in enumerate(['A','B','C','D','E']):
    plt.annotate(txt,(x[i],y[i]),xytext=(5,5),ha='center',textcoords="offset points")
```

```python
# Generating sample dataset  
df = pd.DataFrame({  
    "Age": np.random.randint(20, 60, 50),  
    "Salary": np.random.randint(30000, 100000, 50),  
    "Experience": np.random.randint(1, 20, 50)  
})  

# Scatter plot (Age vs Salary)  
plt.scatter(df["Age"], df["Salary"], c=df["Experience"], cmap="coolwarm", alpha=0.8)  
plt.colorbar(label="Years of Experience")

plt.xlabel("Age")  
plt.ylabel("Salary") 
```
