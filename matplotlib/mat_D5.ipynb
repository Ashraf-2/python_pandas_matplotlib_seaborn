## Subplots

```python
import pandas as pd 
import numpy as nimport matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.figure(figsize=(10,5))

# first sub plot
plt.subplot(1,2,1)
plt.plot(x,y,color='red')
plt.title("Plot 1") 

# second sub plot
plt.subplot(1,2,2)
plt.plot(x,y,color='blue')
plt.title("Plot 2")

plt.suptitle("Sub Plots")
plt.tight_layout()  # Adjust spacing  
plt.show()
```

```python
import numpy as np

fig, ax = plt.subplots(3,2,figsize=(10,5))
x = [1,2,3,4,5]
y = [10,20,30,40,50]

ax[0,0].plot(x,y,color='red')   
ax[0,0].set_title("Line Plot")

ax[0,1].bar(x,y,color='blue')
ax[0,1].set_title("Bar Plot")

ax[1,0].hist(np.random.randn(1000),bins=10, color='green')
ax[1,0].set_title("Histogram")

ax[1,1].scatter(x,y,color='orange')
ax[1,1].set_title("Scatter Plot")


plt.tight_layout() # Adjust spacing
plt.show()
```

```python
import matplotlib.pyplot as plt  

# First figure  
plt.figure(figsize=(6, 4))  
plt.plot([1, 2, 3], [4, 5, 6], color="blue")  
plt.title("Figure 1")  
plt.show()  

# Second figure  
plt.figure(figsize=(6, 4))  
plt.bar(["A", "B", "C"], [10, 15, 7], color="red")  
plt.title("Figure 2")  
plt.show()
```

### High rezulation plot


```python
plt.figure(figsize=(4, 2), dpi=150)  
plt.plot([1, 2, 3], [4, 5, 6], color="blue")  
plt.title("High-Resolution Figure")  
plt.show()
```

```python
import pandas as pd  
import numpy as np  
import matplotlib.pyplot as plt  

# Sample Data  
df = pd.DataFrame({  
    "Month": ["Jan", "Feb", "Mar", "Apr", "May", "Jun"],  
    "Sales": [100, 150, 200, 180, 250, 220],  
    "Profit": [20, 30, 50, 40, 70, 60]  
})  

# Creating subplots  
fig, ax = plt.subplots(1, 2, figsize=(10, 5))  

# Line plot for sales trend  
ax[0].plot(df["Month"], df["Sales"], marker="o", color="blue", label="Sales")  
ax[0].set_title("Sales Trend")  
ax[0].set_xlabel("Month")  
ax[0].set_ylabel("Sales")  
ax[0].legend()  

# Bar chart for profit  
ax[1].bar(df["Month"], df["Profit"], color="green")  
ax[1].set_title("Profit by Month")  
ax[1].set_xlabel("Month")  
ax[1].set_ylabel("Profit")  

# Show  
plt.tight_layout()  
plt.show()
```
