```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Sample Data  
data = {  
    "Month": ["Jan", "Feb", "Mar", "Apr", "May"],  
    "Sales": [100, 120, 140, 160, 180]  
}  
df = pd.DataFrame(data) 
```

```python
df
```

```python
# Plotting the data
plt.figure(figsize=(10, 5))
x = df['Month']
y = df['Sales']
plt.plot(x, y, color='b',linestyle='-', linewidth=2, marker='o', markerfacecolor='r', markersize=8)
plt.title('Monthly Sales')
plt.xlabel('Month')
plt.grid(True)


plt.savefig('line_plot_2.png')
```

```python
# Bar chart
plt.figure(figsize=(10, 5))
plt.bar(x,y, color='b')
plt.title("Sales Comparison")  
plt.ylabel("Sales")  
plt.xlabel("Month")
plt.xticks(rotation=45)  
plt.savefig('Bar chart.png')

# Horizontal Bar chart
plt.figure(figsize=(10, 5))
plt.barh(x,y, color='g',alpha=0.5)
plt.title("Sales Comparison")
plt.ylabel("Month")
plt.xlabel("Sales")
plt.xticks(rotation=45)
plt.savefig('Horizontal Bar chart.png')
 
```

## Seaborn Basics

```python
# random int data
data = np.random.randint(51, 99, 100)
data

#random weather data of 7 days in sylhet
syl_wether ={
    "Day": ["Sun", "Mon", "Tue", "Wed", "Thu", "Fri", "Sat"],
    "Temperature": [30, 32, 33, 31, 32, 30, 31],
    "Humidity": [80, 82, 85, 83, 80, 82, 81],
    "Rain": [0, 1, 0, 0, 1, 0, 0]
}

df = pd.DataFrame(syl_wether)

df
```

### Seaborn Histogram

```python
sns.histplot(data,bins=10, kde=False, color='g', alpha=0.5)
plt.title("Seaborn_Histogram.png")
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.savefig("Seaborn_Histogram.png")
plt.show()
```

```python
sns.histplot(data, bins=15, kde=True, color='r',alpha=0.5, linewidth=2)
plt.title("Seaborn Histogram with KDE") 
plt.xlabel("Data")
plt.ylabel("Frequency")
plt.grid(True)

plt.savefig("seaborn_histogram_KDE.png")

```

### Seaborn Boxplot

```python
df_titanic = sns.load_dataset("titanic")  

df_titanic.info()
```

#### Task: draw a histrogram of male persons and their age in the given dataset

```python
# age of the people who has column df["who"] = "male" in df_titanic
male_age = df_titanic[df_titanic["who"]=="man"]["age"].dropna()
male_age


```

```python
male_age_count = male_age.value_counts().sort_index()
male_age_count
```

```python
# make data frame of male_age_count
df_male_age_count = pd.DataFrame(male_age_count, columns=["count"])
df_male_age_count
```

```python
# now show the data in bar chart
#plt.figure(figsize=(10, 5))
x = df_male_age_count.index
print(x)

y = df_male_age_count["count"]
print(y)
```

```python
# show the bar plot
plt.figure(figsize=(20, 8))
plt.bar(x, y, color='b')
plt.title("Titanic Males Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count/Frequency")
plt.xticks(rotation=45) 

plt.savefig("Titanis_male_age_distribution.png")
plt.show()
```

### Titanic Age Distribution

```python
import seaborn as sns  
import pandas as pd  
import matplotlib.pyplot as plt  

# Load dataset  
df = sns.load_dataset("titanic")  

# Histogram for Age  
sns.histplot(df["age"].dropna(), bins=20, kde=True, color="blue")  
plt.title("Titanic Age Distribution")  
plt.savefig("Titanic_age_distribution.png") 
plt.show()

# Boxplot for Age by Class  
sns.boxplot(x="class", y="age", data=df, palette="coolwarm")  
plt.title("Titanic Age Distribution by Class")  
plt.savefig("Titanic_age_distribution_by_class.png")
plt.show()
```
