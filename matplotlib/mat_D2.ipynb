```python
import pandas as pd
import matplotlib.pyplot as plt
```

## Bar Chart

```python
# Data  
categories = ["A", "B", "C", "D"]  
values = [10, 25, 40, 30]

# plot
plt.bar(categories, values, color='blue', alpha=0.3)
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Bar Chart Example')
plt.show()
```

### Horizontal Bar chart

```python
# Data  
categories = ["A", "B", "C", "D"]  
values = [10, 25, 40, 30]

# plot
plt.barh(categories, values, color='blue', alpha=0.3)
plt.xlabel('Values')
plt.ylabel('Categories')
plt.title('Horizontal Bar Chart Example')
plt.show()
```

### Stacked Bar Chart

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Data  
categories = ["A", "B", "C", "D"]  
values1 = [10, 20, 30, 40]  
values2 = [5, 15, 25, 10]  

plt.bar(categories, values1, label="Group 1", color="blue")  
plt.bar(categories, values2, bottom=values1, label="Group 2", color="orange") 
```

## Histograms 

```python
import numpy as np  
import matplotlib.pyplot as plt  

# Generating random data  
data = np.random.normal(50, 10, 500)  # Mean=50, Std=10, 500 values  

# Plot  
plt.hist(data, bins=10, color="purple", edgecolor="black", alpha=0.7)  

# Customization  
plt.xlabel("Value Range")  
plt.ylabel("Frequency")  
plt.title("Histogram Example")  

# Show  
plt.show()
```

 ## Pie Charts 

### Basic Pie Chart

```python
import matplotlib.pyplot as plt  

# Data  
fruits = ["Apple", "Banana", "Cherry", "Date"]  
quantity = [30, 20, 35, 15]  

# Plot  
plt.pie(quantity, labels=fruits, autopct="%1.1f%%", startangle=50)  

# Customization  
plt.title("Fruit Distribution")  

# Show  
plt.show()
```

###  Exploding Slices & Customizing Colors

```python
import matplotlib.pyplot as plt  

# Data  
fruits = ["Apple", "Banana", "Cherry", "Date"]  
sizes = [30, 20, 35, 15]  
colors = ["red", "yellow", "pink", "brown"]  
explode = (0, 0.2, 0, 0.2)  # Explode the second slice  

# Plot  
plt.pie(sizes, labels=fruits, autopct="%1.1f%%", colors=colors, explode=explode, startangle=140, shadow=True)  

# Customization  
plt.title("Exploded Pie Chart")  

# Show  
plt.show()
```
