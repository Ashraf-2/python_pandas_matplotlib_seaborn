```python
import pandas as pd
```

```python
# Define data for the DataFrame
data = {
    "Product Name": ["Wireless Mouse", "Bluetooth Headphones", "4K Monitor", "Smartphone", "USB-C Cable"],
    "Category": ["Electronics", "Electronics", "Electronics", "Mobile", "Accessories"],
    "Price ($)": [25.99, 79.99, 299.99, 699.99, 15.99],
    "Stock": [120, 80, 50, 200, 500],
    "Rating": [4.5, 4.2, 4.7, 4.6, 4.1]
}

# Create DataFrame
df = pd.DataFrame(data)

df
```

```python
# Set 'Name' column as the index
df_with_index = df.set_index('Product Name')
```

```python
# get the second row in a dataframe view
df.loc[[2,]]
```

```python
# get the category name of the product of second row.
df.loc[2, 'Category']
```
